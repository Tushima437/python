## Traduciendo Columas y Variables

```python
datos.columns
```

```python
##Creo diccionario
dic_columnas = {
    'LIMIT_BAL' : 'limite',
    'CHECKING_ACCOUNT': 'cuenta_corriente',
    'EDUCATION':'escolaridad',
    'MARRIAGE':'estado_civil',
    'AGE':'edad',
    'BILL_AMT':'valor_factura',
    'PAY_AMT':'valor_pago',
    'DEFAULT':'moroso'
}
```

```python
tarjetas = datos.rename(columns = dic_columnas)
```

```python
tarjetas.head()
```

```python
tarjetas.cuenta_corriente.unique()
```

```python
dic_cuenta_c = {
    'Yes': 'Si',
    'No' : 'No'
}
```

```python
tarjetas.cuenta_corriente.map(dic_cuenta_c)
```

```python
tarjetas.cuenta_corriente = tarjetas.cuenta_corriente.map(dic_cuenta_c)
tarjetas.head()
```

```python
tarjetas.escolaridad.unique()
```

```python
dic_escolaridad = {
    '2.University':'2.Universidad',
    '3.Graduate School': '3.Posgrado',
    '1.High School':'1.Secundaria'
}
```

```python
tarjetas.escolaridad = tarjetas.escolaridad.map(dic_escolaridad)
tarjetas.head()
```

```python
tarjetas.estado_civil.unique()
```

```python
dic_estado_civil = {
    'Married' : 'Casado/a',
    'Single'  :  'Soltero/a'
}
```

```python
tarjetas.estado_civil=tarjetas.estado_civil.map(dic_estado_civil)
tarjetas.head()
```
