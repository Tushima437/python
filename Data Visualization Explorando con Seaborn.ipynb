##Test de hipotesis

```python
sns.lmplot(x='iu',y='valor_factura', data=tarjetas, col='moroso')
```

```python
from scipy.stats import ranksums
```

H<sup>null<sup>

La distribucion de los grupos morosos y no morosos es la misma

H<sup>alt<sup>

La distribucion de los grupos morosos y no morosos no es la misma

```python
moroso =tarjetas.query('moroso == 1').valor_factura
moroso
```

```python
no_moroso =tarjetas.query('moroso == 0').valor_factura
no_moroso
```

```python
ranksums(moroso, no_moroso)
#generalmente Valores abajo del 5 ya sirven para aceptar la hipotesis alternativa
```

```python
resultado = ranksums(moroso, no_moroso)
```

```python
print("el p_value de nuestro test es {}".format(resultado.pvalue))
```

Teniendo en cuenta esto, ¿Cómo podemos definir un Pairplot?


-Una función que comparará las variables categóricas en nuestra base de datos y generará los gráficos que mejor relación presenten.


Alternativa incorrecta, un Pairplot no analizará las variables categóricas, esta función analizará las variables numéricas.



-Una imagen para representar gráficamente los promedios de las variables numéricas en nuestra base de datos.


Alternativa incorrecta, un Pairplot no creará un sólo gráfico (esto sólo ocurrirá en el caso de que tengamos una sola variable numérica), creará un número de columnas y filas igual al número de variables numéricas en nuestra base. No mostrará los promedios, nos mostrará un histograma cuando está comparando una variable con ella misma y un gráfico de dispersión cuando es comparada con otras variables.



-Una función que creará una cuadrícula de gráficos de forma que cada variable numérica será analizada con las demás variables numéricas de la base de datos y con ella misma.


Alternativa correcta, un Pairplot creará una columna y una fila para cada variable numérica presente en nuestra base de datos y hará un análisis de cada variable individualmente y como se relaciona con las demás variables numéricas de la base.
