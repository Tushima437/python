#Selecciones y frecuencias

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial.csv', sep= ';')
datos.head(10)
```

```python
# Seleccione solamente los inmuebles clasificados con tipo 'Departamento'
seleccion = datos['Tipo'] == 'Departamento'
n1 = datos[seleccion].shape[0]
n1
```

```python
# Seleccione los inmuebles clasificados con tipos 'Casa', 'Casa en condominio' y 'Casa de villa'
seleccion = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
n2 = datos[seleccion].shape[0]
n2
```

```python
# Seleccione los inmuebles con área entre 60 y 100 metros cuadrados, incluyendo los limites
# 60 <= Area <= 100
seleccion = (datos['Area'] >= 60) & (datos['Area'] <= 100)
n3 = datos[seleccion].shape[0]
n3
```

```python
# Seleccione los inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00
seleccion = (datos['Cuartos'] >= 4) & (datos['Valor'] < 2000)
n4 = datos[seleccion].shape[0]
n4
```

```python
print("Nº de inmuebles clasificados con tipo 'Departamento' -> {}".format(n1))
print("Nº de inmuebles clasificados con tipos 'Casa', 'Casa en condominio' y 'Casa de villa' -> {}".format(n2))
print("Nº de inmuebles con área entre 60 y 100 metros cuadrados, incluyendo los limites -> {}".format(n3))
print("Nº de inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00 -> {}".format(n4))
```

```python
#Utilizando los recursos aprendidos en esta aula, juzgue el conjunto de operaciones de limpieza del DataFrame a continuación:
inmuebles= pd.DataFrame([['Departamento', None, 970, 68],
                        ['Departamento', 2000, 878, 112],
                        ['Casa', 5000, None, 500],
                        ['Departamento', None, 1010, 170],
                        ['Departamento', 1500, 850, None],
                        ['Casa', None, None, None],
                        ['Departamento', 2000, 878, None],
                        ['Departamento', 1550, None, 228],
                        ['Departamento', 2500, 880, 195]],
                        columns = ['Tipo', 'Valor', 'Mantenimiento', 'Impuesto'])
inmuebles
```

```python
#1) Elimina los registros que tengan la columna Valor nula:
inmuebles.dropna(subset = ['Valor'], inplace = True)
inmuebles
```

```python
#4) Reconstruye el índice del DataFrame resultante:
inmuebles.index = range(inmuebles.shape[0])
inmuebles
```

#Metodos de importacion

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
s
```

```python
s.fillna(0)
```

```python
s #Sigue igual por que no hay inplace = True :v
```

```python
s.fillna(method='ffill')  #Cambia el null por el anterior, si el primero es null se deja null
```

```python
#DE PRUEBA
# Si el primer valor es un NaN el ffill lo deja igual :v
data2 = [None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s2 = pd.Series(data2)
s2.fillna(method='ffill')
```

```python
s.fillna(method='bfill') #Igual que ffill pero de abajo pa arriba
```

```python
s.fillna(s.mean()) #s.mean() saca la media
```

```python
s.fillna(method='ffill', limit = 1) #limite de veces que lo repite
```

```python
s1 = s.fillna(method='ffill', limit = 1)  #completo nulos de adelante al final 1 vez
s1
```

```python
s2 = s1.fillna(method='bfill', limit = 1) #termino de completar pero con bfill,
s2
```
